```python
# Import necessary libraries
import requests
import json

# Define the base URL of the API
BASE_URL = "https://opendata.swiss/api/3/action/package_search"

# Define query parameters to search for datasets tagged with "economy"
params = {
    "fq": "tags:economy"
}

# Make a GET request to the API
response = requests.get(BASE_URL, params=params)

# Check if the request was successful
if response.status_code == 200:
    data = response.json()
    print(json.dumps(data, indent=4))  # Pretty-print the JSON response
else:
    print(f"Error {response.status_code}: Unable to fetch data")
```

```python
# Get a list of all organizations
org_url = "https://ckan.opendata.swiss/api/3/action/organization_list"
org_response = requests.get(org_url).json()

# Display first 10 organizations
print(org_response["result"][:10])
```

```python
# Get a list of all datasets
dataset_url = "https://opendata.swiss/api/3/action/package_list"
dataset_response = requests.get(dataset_url).json()

# Display the first 10 datasets
print(dataset_response["result"][:10])
```

```python
# Get datasets from Bundesamt für Statistik (BFS)
bfs_url = "https://ckan.opendata.swiss/api/3/action/package_search?fq=organization:bundesamt-fur-statistik-bfs"
bfs_response = requests.get(bfs_url).json()

# Extract dataset titles
bfs_datasets = [dataset["title"] for dataset in bfs_response["result"]["results"]]

# Display first 5 BFS datasets
print(bfs_datasets[:5])
```

```python
# Get most common tags
tags_url = "https://ckan.opendata.swiss/api/3/action/package_search?facet.field=[%22tags%22]&facet.limit=10&rows=0"
tags_response = requests.get(tags_url).json()

# Display the most frequent tags
print(tags_response)
```

```python
# Search keywords sorted by relevance
keywords_url = "https://ckan.opendata.swiss/api/3/action/package_search?facet.field=[%22keywords%22]&sort=relevance+asc"
keywords_response = requests.get(keywords_url).json()

# Display sorted keyword data
print(json.dumps(keywords_response, indent=4))
```
